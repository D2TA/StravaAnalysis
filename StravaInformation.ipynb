#Strava Data Extraction Process:

```
import numpy as np
import pandas as pd
import sklearn as sk
import seaborn as sns
import random 
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings("ignore") #warning suppressor

from matplotlib import pyplot as plt
pd.options.display.max_columns=500

## Reading Datasets - Activities
StravaActivity = pd.read_csv('activities.csv')
StravaActivity.tail()
```

```
## List/Set of Activity Dates
SetActivityDate = set(StravaActivity['Activity Date'])
## Number of Activities
print(f'There are {len(SetActivityDate)} number of activities')
SetActivityDay = (pd.to_datetime(StravaActivity['Activity Date']).dt.date).unique()
## How many days of activites are there? any duplicate dates?
print(f'From {SetActivityDay.min()} to {SetActivityDay.max()}, you cycled on {len(SetActivityDay)} different days')

## How many bikes does our cyclist have?
SetActivityGear = set(StravaActivity['Activity Gear'])
print(f'Our cyclist has riden on {len(SetActivityGear)} different types of bikes: {SetActivityGear}')

## What is the Total Distance Riden?
TotalDistanceKM = round(StravaActivity['Distance'].sum(),2)
TotalDistance = round(StravaActivity['Distance.1'].sum(),2)
print(f'The total distance riden is { TotalDistanceKM } (km)')

## What is the average Distance per ride riden?
AverageDistanceKM = round(StravaActivity['Distance'].mean(),2)
AverageDistance = round(StravaActivity['Distance.1'].mean(),2)
print(f'The average distance per ride is {AverageDistanceKM} (km)')

## What is the Total Time Riden (Moving Time)?
TotalTimeHR = round(StravaActivity['Moving Time'].sum()/3600,2)
print(f'The total time riden is {TotalTimeHR} (hours)')

## What is the average riding time?
AverageTimeHR = round(StravaActivity['Moving Time'].mean()/3600,2)
print(f'The average time riden is {AverageTimeHR} hours of cycling')
print(f'The average time riden is {AverageTimeHR*60} minutes of cycling')

## For rides with a heart rate monitor, What was the average heart rate?
AverageBPM = round(StravaActivity['Average Heart Rate'].mean(),2)
print(f'The average heart beat per minute is {AverageBPM} during cycling')

## For rides, what is the average calories burned?
AverageCal = round(StravaActivity['Calories'].mean(),2)
print(f'The average calories burned is {AverageCal} during cycling')
```

#Data Conversion
1.   Activity Date is in Date Time: Parse to Date
2.   Activity Date is in Date Time: Parse to Time
3.   Acitivity Gear is in Categorical: 
  *   0 : Marin Lombard 1
  *   1 : Aventon Mataro




```
print('Performing Data Transformation - Date Time Field Split : Start')
## Datetime Conversion
StravaActivity['ActivityDate'] = pd.to_datetime(StravaActivity['Activity Date']).dt.date
StravaActivity['ActivityTime'] = pd.to_datetime(StravaActivity['Activity Date']).dt.time

print('Performing Data Transformation - Categorical Data : Start')
## Categorical Gear Conversion
StravaActivity['ActivityGearID'] = np.where(StravaActivity['Activity Gear'] == 'Marin Lombard 1', 0, 1)

print('Performing Data Transformation : Completed')
```

# Distance Visualization

```
## Distance Data Preperation
distancePerDay = pd.DataFrame(data = StravaActivity[['ActivityDate','Distance.1']])
distancePerDay = distancePerDay.groupby(by = ['ActivityDate']).sum()/1000
distancePerDay['Cumulative Distance'] = distancePerDay.cumsum()
distancePerDay['Rolling 6 Week Average Distance'] = distancePerDay['Distance.1'].rolling(window = 6).mean()
```

```
## Distance Visualization
distanceFig = make_subplots(specs = [[{'secondary_y': True}]])
  ## Add Traces
distanceFig.add_trace(
    go.Scatter(x = distancePerDay.index, y = distancePerDay[distancePerDay.columns[1]], name = distancePerDay.columns[1])
    , secondary_y = False
)

distanceFig.add_trace(
    go.Scatter(x = distancePerDay.index, y = distancePerDay[distancePerDay.columns[0]], name = 'Daily Distance Achieved')
    , secondary_y = True
)

distanceFig.add_trace(
    go.Scatter(x = distancePerDay.index, y = distancePerDay[distancePerDay.columns[2]], name = distancePerDay.columns[2]
    , line = dict(color = 'royalblue', width = 2, dash = 'dash'))
    , secondary_y = True
)

distanceFig.update_layout(
    title_text = f'Cycling activity from {SetActivityDay.min()} to {SetActivityDay.max()}:'
)

distanceFig.update_xaxes(
    tickformat = '%Y-%m-%d'
    , title = 'Daily Cycling Activity'
)
distanceFig.update_yaxes(
    title = '<b> Total <b> Distance (in km)'
    , secondary_y = False
)
distanceFig.update_yaxes(
    title = '<b> Daily <b> Distance (in km)'
    , secondary_y = True
)


distanceFig.show()
```

#Heart Rate Visualization

```
## Heart Rate Data Preperation
AverageHeartRatePerDay = pd.DataFrame(data = StravaActivity[['ActivityDate','Average Heart Rate']])
AverageHeartRatePerDay = AverageHeartRatePerDay.dropna()
AverageHeartRatePerDay = AverageHeartRatePerDay.groupby(by = AverageHeartRatePerDay['ActivityDate']).mean()

distancePerDay = distancePerDay.merge(AverageHeartRatePerDay, left_index=True, right_index=True)
```

```
## Heart Rate Visualization
HeartRateFig = px.scatter(distancePerDay, x = distancePerDay.index, y = distancePerDay.columns[3]
                          , color = distancePerDay.columns[3]  
                          , size = distancePerDay.columns[0]
                          , color_continuous_scale = 'redor'
)

HeartRateFig.update_traces(marker = 
                           dict(
                               line = dict(width = 2
                                           , color = 'purple')
                               )
)

HeartRateFig.show()
```
